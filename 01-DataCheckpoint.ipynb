## Research Question

How well geographic and demographic factors predict life expectancy between first, second and third world countries across decades from 1960 to today. Do the most important factors change over time? Do some factors lose predictive weight as nations transition between economic tiers?


## Background and Prior Work

**Updated**\
Since the 1950's human life expectancy, has skyrocketed. Most people today have higher life expectancies than the richest countries in 1950 <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). Global health inequality has also diminished significantly since the 1950s; a newborn in some countries in the 1950s could expect to live twice as much as a newborn in other countries. An interesting factor to note is that a big factor in contributing to these higher life expectancies is the higher infant mortality rates in the 1950s that skews the number a bit, however, life expectancy has still increased for all ages regardless <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2).

There exists a lot of prior research on topics of what has helped increase global life expectancy. A study in the Jounrnal of Economics identifies factors that contributed to this as improved nutrition, better public health through government, urbanization, vaccination, and medical improvements. Furthermore, since WW2 (1945), poor countries have gained rapid improvements in life expectancy, with some gaining over one year per year <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). It's interesting to note that this may imply life expectancy correlates with diminishing returns, where a lot of gains can be made from basic quaility of life changes, however it starts plateuing at the top. 

A useful graph to track changes across time is called the Preston Curve. This graph plots life expectancy against GDP. Over time, this curve has been observed to have risen, which matches with what we expect given the higher life expectancy. However, it is equally important to note that while this trend holds as a large, individual countries have been observed displaying the opposite trend at times as well. <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). This brings up an important distinction that while income is a good predictor of health, there are still other factors at play in determining life expectancy.

Furthermore, while the Preston Curve illustrates that GDP is a foundational predictor, its flattening at higher income levels suggests a transition in the underlying drivers of longevity. This project aims to move beyond aggregate economic metrics to identify which specific geographic and demographic factors (such as urbanization rates, education levels, or climate stability) gain predictive weight as nations move up the economic tier. By analyzing these shifts across decades, we seek to determine if the "formula" for increasing life expectancy has fundamentally changed, and whether certain variables that were critical in a 1960s "Third World" context have lost their predictive power in the modern era.

When countries transition into a different economic tier, it can be broadly attributed to a couple of factors using a long time scale. These factors are energy use, trade, capital, labour, human capital development and foreign direct investment <a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note5).However interestingly enough in a short time frame, human capitial has a negative impact, while energy usage and foreign direct investment have a positive impact. <a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note5). 

And whenever countries shift economic tiers and grow towards the top of the Preston Curve, it can be seen that the factors that affected life expectancy the most stop having such a huge effect while other factors grow in importance. This can be seen with the shift in effect of certain diseases. In developing countries, infectious diseases is what affects life expectancy the most through other factors like access to clean water. But once countries develop more, they increase access to these factors that decreases risk of infectious disease so more people develop chronic illness which then affect thier life expectancy <a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note6). As countries shift on the Preston Curve, different factors lose their weight on life expectancy and other factors replace them.

Life Expectancy: 
This dataset focuses on global life expectancy metrics over time, specifically tracking annual changes and gender-based disparities. It contains 17191 observations across several key variables (after dropping empty columns like region and income_group). The data is geographical, allowing for an analysis of how longevity shifts year-over-year across different countries.

We had to wrangle a lot of the data and include values that were 1.5 standard deviations within the mean. Since there were certain outliers like Central African Republic were over 20+ standard deviations from the mean due to missing data, we thought that would negatively impact the data. After dropping observations that were over 1.5 standard deviations away and dropping null values, we ended with ~16000 observations.

```python
# imports
import pandas as pd
import seaborn as sns
```

```python
# Data is already tidy
life_df = pd.read_csv("./data/00-raw/life_expectancy.csv")
life_df
```

```python
# Size of dataset
life_df.shape
```

```python
# Region, income group and iso2_code are missing for all entries
life_df.isna().sum()
```

```python
life_df = life_df.drop(columns=["region", "income_group", "iso2_code", "country_code"])
```

```python
# le_total_yoy_change has some very large outliers compared to the standard deviation
life_df.describe()
```

```python
# There are many entries with le_total_yoy_change over 30 std away
life_df.sort_values("le_total_yoy_change")
```

```python
# Removing outlier year changes (keeping within 1.5 std)
std = life_df.describe()["le_total_yoy_change"]["std"]
life_df = life_df[life_df["le_total_yoy_change"].abs() < 1.5 * std]
```

```python
sns.pairplot(life_df)
```

```python
life_df.isna().sum() # removing outliers in life expectancy got rid of null entries as well
```
